```python
def fac(n):
    if n == 1:
        return 1
    
    return n * fac(n-1)
    
```

```python
print(fac(100))
```

```python
def fac_mem(n, m):
    if n in m:
        return m[n]
    
    if n == 1:
        return 1
    
    m[n] = n * fac(n-1)
    
    return m[n]
```

```python
print(fac_mem(100, {}))
```

```python
def fib(n):
    if n == 1 or n == 2:
        return 1
    
    return fib(n-1) + fib(n-2)
```

```python
print(fib(35))
```

```python
def fib_mem(n,m):
    
    if n == 1 or n == 2:
        return 1
    
    if n in m:
        return m[n]
    
    m[n] = fib_mem(n-2,m) + fib_mem(n-1,m)
    
    return m[n] 
```

```python
print(fib_mem(35, {}))
```

```python
def steps_mem(n, x, m):

    if n == 0:
        return 1
    if n < 0:
        return 0
    
    if n in m:
        return m[n]

    for e in x:
        if n in m:
            m[n] += steps_mem(n - e, x, m)
        else:
            m[n] = steps_mem(n - e, x, m)
    print(m)
    return m[n]
```

```python
steps_mem(6, [4,5], {})
```

```python
def steps(n, x):

    if n == 0:
        return 1
    if n < 0:
        return 0
    
    total = 0
    for e in x:

        total += steps(n - e, x)
  
    return total
```

```python
steps(35, [1,3,5])
```

```python
def lastStoneWeight(stones) -> int:

    a = sorted(stones)
    while len(a) > 1:
        x = a.pop()
        y = a.pop()

        if x != y:
            a.append(x-y)
            
        a = sorted(a)

    if len(a) == 1:
        return a[0]
    return 0
```

```python
lastStoneWeight([1,1,1])
```
